### EDA on the Chicago Crime Data

What is Exploratory Data Analysis?

Exploratory Data Analysis (EDA), also known as Data Exploration, is a step in the Data Analysis Process, where a number of techniques are used to better understand the dataset being used.

‘Understanding the dataset’ can refer to a number of things including but not limited to…

Extracting important variables and leaving behind useless variables Identifying outliers, missing values, or human error Understanding the relationship(s), or lack of, between variables

Ultimately, maximizing your insights of a dataset and minimizing potential error that may occur later in the process Exploratory Data Analysis (EDA) is like the Sherlock Holmes of data science. It involves delving deep into datasets, using visual methods as its magnifying glass, to highlight significant attributes. EDA helps us unearth hidden patterns, relationships, and trends within the data before we even start modeling.

Let’s dive into Exploratory Data Analysis (EDA) using Chicago crime data as our case study. We’ll leverage Pandas to explore and summarize the dataset. Here are the steps we’ll follow

### Filling Missing Values

```python
df['Case Number']
```

```python
# Fill missing values in 'Case Number' column with mode
mode_case_number = df['Case Number'].mode()[0]
df['Case Number'].fillna(mode_case_number, inplace=True)
```

```python
# Check if missing values are filled
print(df['Case Number'].isnull().sum())
```

```python
df['Location Description']
```

```python
df['Location Description'].mode()
```

```python
# Fill missing values in 'Location Description' column with mode
df['Location Description'] = df['Location Description'].fillna(df['Location Description'].mode()[0])
```

```python
# Check if missing values are filled
print(df['Location Description'].isnull().sum())
```

```python
df['District']
```

```python
#Filling the missing value in 'District' with backward fill

df['District'].fillna(method='bfill', inplace=True)
```

```python
# Check if missing values are filled
print(df['District'].isnull().sum())
```

```python
df['Ward']
```

```python
# Fill missing values in 'Ward' column with linear interpolation
df['Ward'] = df['Ward'].interpolate(method='linear')
```

```python
# Check if missing values are filled
print(df['Ward'].isnull().sum())
```

```python
#Filling the missing value in 'Community Area' with forward fill
df['Community Area'].fillna(method='ffill', inplace=True)
```

```python
# Check if missing values are filled
print(df['Community Area'].isnull().sum())
```

```python
df['X Coordinate']
```

```python
# Fill missing 'X Coordinates' values with mean 
# This method assumes that the missing values are missing at random and that 
# the distribution of X Coordinates values is approximately normal.

mean_xcoordinates = df['X Coordinate'].mean()
df['X Coordinate'].fillna(mean_xcoordinates, inplace=True)
```

```python
# Check if missing values are filled
print(df['X Coordinate'].isnull().sum())
```

```python
df['Y Coordinate']
```

```python
# Fill missing 'Y Coordinates' values with mean 
# This method assumes that the missing values are missing at random and that 
# the distribution of Y Coordinates values is approximately normal.

mean_ycoordinates = df['Y Coordinate'].mean()
df['Y Coordinate'].fillna(mean_ycoordinates, inplace=True)
```

```python
# Check if missing values are filled
print(df['Y Coordinate'].isnull().sum())
```

```python
df['Latitude']
```

```python
# Fill missing latitude values with mean
mean_latitude = df['Latitude'].mean()
df['Latitude'].fillna(mean_latitude, inplace=True)
```

```python
# Check if missing values are filled
print(df['Latitude'].isnull().sum())
```

```python
df['Longitude']
```

```python
# Fill missing longitude values with mean
mean_longitude = df['Longitude'].mean()
df['Longitude'].fillna(mean_longitude, inplace=True)
```

```python
# Check if missing values are filled
print(df['Longitude'].isnull().sum())
```

```python
# filling missing location values
# Check the data type of the 'Location' column
print(df['Location'].dtype)
```

```python
# Noticed that the Location column is stored as an object type
# Example of the 'Location' column format: '(41.8578, -87.6612)'
# Extract latitude and longitude from the 'Location' column
df[['Latitude', 'Longitude']] = df['Location'].str.extract(r'\(([^,]+), ([^)]+)\)').astype(float)

# Fill missing values for latitude and longitude
df['Latitude'].fillna(df['Latitude'].mean(), inplace=True)
df['Longitude'].fillna(df['Longitude'].mean(), inplace=True)

# Combine latitude and longitude back into the original format
df['Location'] = '(' + df['Latitude'].astype(str) + ', ' + df['Longitude'].astype(str) + ')'

```

```python
# Check if missing values are filled
print(df['Location'].isnull().sum())
```

```python
# Check for missing values
df.isna().sum()
```

## Feature Engineering

Use functions to create new columns for Months, Day, Season

```python
#inspecting the date column 
(df['Date'].head()) 
```

```python
# Function to extract and map months
def extract_and_map_month(date_column):
    # Convert 'Date' column to datetime
    date_column = pd.to_datetime(date_column)
    
    # Extract month
    month = date_column.dt.month
    
    # use a dictionary to map month numbers to month names
    month_map = {
        1: 'January',
        2: 'February',
        3: 'March',
        4: 'April',
        5: 'May',
        6: 'June',
        7: 'July',
        8: 'August',
        9: 'September',
        10: 'October',
        11: 'November',
        12: 'December',
        
        
    }
    
    return month.map(month_map)

# Add a new column 'Month' with mapped month values
df['Month'] = extract_and_map_month(df['Date']) 
```

```python
 # display the updated column with months of the year
df.head()
```

```python
# Function to extract and map days of the week
def extract_and_map_days(date_column):
    # Convert 'Date' column to datetime
    date_column = pd.to_datetime(date_column)
    
    # Extract day of the week (0=Monday, 6=Sunday)
    days = date_column.dt.dayofweek
    
    # Map day of the week to day names
    day_map = {
        0: 'Monday',
        1: 'Tuesday',
        2: 'Wednesday',
        3: 'Thursday',
        4: 'Friday',
        5: 'Saturday',
        6: 'Sunday'
    }
    
    return days.map(day_map)

# Add a new column 'Days' with mapped day values
df['Days'] = extract_and_map_days(df['Date'])
```

```python
# display the updated column with Days of the Week
df.head(5) 
```

```python
# Creating a column for Seasons
def create_seasons_column(date_column):
    """
    Creates a new column representing seasons based on the months in a datetime column.

    Parameters:
    - date_column: Pandas Series, representing a datetime feature

    Returns:
    - Pandas Series containing the corresponding season names
    """
    # Ensure the input is in datetime format
    date_column = pd.to_datetime(date_column, format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

    # Define a mapping of month to season
    month_to_season = {
        1: 'Winter', 2: 'Winter', 3: 'Spring',
        4: 'Spring', 5: 'Spring', 6: 'Summer',
        7: 'Summer', 8: 'Summer', 9: 'Fall',
        10: 'Fall', 11: 'Fall', 12: 'Winter'
    }

    # Extract the month and map to season
    seasons_column = date_column.dt.month.map(month_to_season)

    return seasons_column
```

```python
df['Seasons'] = create_seasons_column(df['Date'])
```

```python
# display the updated column with Seasons of the year
df.head()
```

## Univariate Analysis

Summary Statistics

```python
df.describe()
```

Crime Types Distribution

```python
# Plot the distribution of crime types
plt.figure(figsize=(10, 6))
sns.countplot(y='Primary Type', data=df, order=df['Primary Type'].value_counts().index)
plt.title('Distribution of Crime Types')
plt.xlabel('Count')
plt.ylabel('Crime Type')
plt.show()
```

 Location Distribution

```python
# Plot the distribution of crime locations
plt.figure(figsize=(10, 6))
sns.countplot(y='Location Description', data=df, order=df['Location Description'].value_counts().iloc[:10].index)
plt.title('Top 10 Crime Locations')
plt.xlabel('Count')
plt.ylabel('Location')
plt.show()
```

## Multivariate Analysis

Crime Trends Over Time

```python
# Plot the number of crimes over time
plt.figure(figsize=(12, 6))
df.groupby('Year')['ID'].count().plot(marker='o')
plt.title('Number of Crimes Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.grid(True)
plt.show()
```

Crime Types by Location

```python
# Plot the distribution of crime types by location
plt.figure(figsize=(12, 8))
crime_location = df.groupby(['Location Description', 'Primary Type']).size().unstack()
crime_location.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Crime Types by Location')
plt.xlabel('Location')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=90)
plt.legend(title='Primary Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## Correlation Analysis

### Heatmap of Correlation Matrix


```python
# Visualize the correlation between different numerical variables in the dataset using a heatmap.
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
